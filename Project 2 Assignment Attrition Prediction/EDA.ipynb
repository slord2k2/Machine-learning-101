```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True,fmt='.2f',cmap='magma')
plt.title('Correlation Heatmap',fontsize=35)
```

From the above Heat Map we can conclude that Monthly income and Job level can replace each other and years in Current Role and year since last Promotion is underlying information as welll as years with current Manager.

```python
sns.countplot(x='Attrition', data=df)

# Add labels and title
plt.title('Count of Attrition', fontsize=30)
plt.xlabel('Attrition')
plt.ylabel('Count')

ax = plt.gca()
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
```

From the above graph we can see ,observations belonging to the ‘No’ category is way greater than that belonging to ‘Yes’ category. Hence we have skewed classes and this is a typical example of the ‘Imbalanced Classification Problem’.

```python
size = df['BusinessTravel'].value_counts()
labels = df['BusinessTravel'].unique()
colors = plt.cm.Wistia(np.linspace(0,1,5))

plt.figure(figsize=(10,10))
circle= plt.Circle((0,0),radius=0.7,color='white')
plt.pie(size,colors=colors , labels=labels , shadow=True, autopct='%0.2f%%')
p=plt.gcf()
p.gca().add_artist(circle)
plt.title("% of Business Travel",fontsize=30)
plt.legend(bbox_to_anochor=(0.5,0,0.9,0.9),loc='upper left')
```

```python
plt.figure(figsize=(12,6))
sns.distplot(df['Age'], bins=30, hist_kws=dict(edgecolor='black', linewidth=2))
plt.title('Distribution of Age of the Employees', fontsize=30)
plt.xlabel('Age')
```

```python
plt.figure(figsize=(17,6))
sns.countplot(x='Age', hue='Attrition', data=df)
plt.title('Dependency of Age in determining Attrition', fontsize=30)
plt.xlabel('Age')
```

the attrition is maximum between the age groups 28–32. The attrition rate keeps on falling with increasing age, as people look after stability in their jobs at these point of times. Also at a very younger age, i.e. from 18–20, the chances of an employee leaving the organization is far more- since they are exploring at that point of time. It reaches a break even point at the age of 21

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Attrition', y='Age', hue='Gender',data=df)
plt.title('Attrition wrt Age and Gender', fontsize=30)
plt.xlabel('Attrition')
```

Note that the attrition mean age is slightly less than the other group.

```python
data=pd.crosstab(df['BusinessTravel'],df['Attrition'])
data.div(data.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,color=['brown', 'cyan'],figsize=(12,6))
plt.title("Dependency of OverTime on Attrition",fontsize=30)
plt.xlabel('OverTime')
```

Travel Frequency has a huge impact on the attrition of employees

```python
data=pd.crosstab(df['Department'],df['Attrition'])
data.div(data.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,color=['orange', 'magenta'],figsize=(12,6))
plt.title("Dependency of Department on Attrition",fontsize=30)
plt.xlabel('Department')
```

```python
data=pd.crosstab(df['JobRole'],df['Attrition'])
data.div(data.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,color=['orange', 'black'],figsize=(12,6))
plt.title("Dependency of JobRole on Attrition",fontsize=30)
plt.xlabel('JobRole')
```

Attrition rate is highest in Sales followed by the HR and Lab Technician

```python
data=pd.crosstab(df['OverTime'],df['Attrition'])
data.div(data.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,color=['brown', 'black'],figsize=(12,6))
plt.title("Dependency of OverTime on Attrition",fontsize=30)
plt.xlabel('OverTime')
```
